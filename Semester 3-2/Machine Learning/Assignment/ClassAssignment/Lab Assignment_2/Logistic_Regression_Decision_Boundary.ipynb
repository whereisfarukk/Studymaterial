```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
import matplotlib as mpl
```

```python
pts = np.loadtxt('/content/inpts.txt')
```

```python
X = pts[:,:2]
y = pts[:,2].astype('int')
```

```python
model = sklearn.linear_model.LogisticRegression()
model.fit(X, y)
```

```python
b = model.intercept_[0]
w1, w2 = model.coef_.T
b, w1, w2
```

```python
c = -b/w2
m = -w1/w2
c, m
```

```python
xmin, xmax = -1, 2
ymin, ymax = -1, 2.5
```

```python
xd = np.array([xmin, xmax])
yd = m*xd + c
```

```python
plt.plot(xd, yd, 'k', lw=1, ls='--')
plt.fill_between(xd, yd, ymin, color='red', alpha=0.1)
plt.fill_between(xd, yd, ymax, color='blue', alpha=0.1)

plt.scatter(*X[y==0].T, s=10)
plt.scatter(*X[y==1].T, s=10)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.ylabel(r'x2')
plt.xlabel(r'x1')

plt.show()
```
